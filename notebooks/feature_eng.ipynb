```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
data = pd.read_csv("../data/processed/car_data.csv")
```

```python
seller_type_encode = LabelEncoder()
data['seller_type'] = seller_type_encode.fit_transform(data['seller_type'])
data.head()
```

```python
fuel_type_encode = LabelEncoder()
data['fuel_type'] = fuel_type_encode.fit_transform(data['fuel_type'])
data.head()
```

```python
transmission_type_encode = LabelEncoder()
data['transmission_type'] = transmission_type_encode.fit_transform(data['transmission_type'])
data.head()
```

```python
brand_encode = LabelEncoder()
data['brand'] = brand_encode.fit_transform(data['brand'])
data.head()
```

```python
model_encode = LabelEncoder()
data['model'] = model_encode.fit_transform(data['model'])
data.head()
```

```python
import joblib

def save_encoders(encoders: dict, path_prefix: str):
    for name, encoder in encoders.items():
        joblib.dump(encoder, f"../models/encoders/{path_prefix}_{name}.pkl")
```

```python
encoders = {
    "seller_type" : seller_type_encode,
    "fuel_type": fuel_type_encode,
    "transmission_type" : transmission_type_encode,
    "brand" : brand_encode,
    "model": model_encode
}

save_encoders(encoders, "encoder")
```

```python
data.to_csv("../data/processed/data.csv",index=False)
```
